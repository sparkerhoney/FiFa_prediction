```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
''' 
 id : 선수 고유의 아이디  -> 필요없음
 name : 이름              -> 필요없음
 age : 나이
 continent : 선수들의 국적이 포함되어 있는 대륙입니다
 contract_until : 선수의 계약기간이 언제까지인지 나타내어 줍니다
 position : 선수가 선호하는 포지션입니다. ex) 공격수, 수비수 등
 prefer_foot : 선수가 선호하는 발입니다. ex) 오른발
 reputation : 선수가 유명한 정도입니다. ex) 높은 수치일 수록 유명한 선수
 stat_overall : 선수의 현재 능력치 입니다.
 stat_potential : 선수가 경험 및 노력을 통해 발전할 수 있는 정도입니다.
 stat_skill_moves : 선수의 개인기 능력치 입니다.
 value : FIFA가 선정한 선수의 이적 시장 가격 (단위 : 유로) 입니다
 '''
```

```python
# 불필요한 컬럼 제거
train.drop(['id', 'name'], axis = 1, inplace = True)
test.drop(['id', 'name'], axis = 1, inplace = True)

train.head()
```

```python
#contract_until 변수 int 형으로 변환
# 계약 연도만 추출
def func(string:object) -> int:
    """계약 연도만 추출하여 int로 반환"""
    string = string[-4:]
    return int(string)


train['contract_until'] = train['contract_until'].apply(func)
test['contract_until'] = test['contract_until'].apply(func)
```

# 1변수 , 나이별 평판이 관계가 있는 것으로 나타남 

```python
from sklearn.model_selection import train_test_split #랜덤 분할, 트레인 테스트 분리하기위한 함수 불러오기  A안에 B를 불러오겠다

train_set, test_set = train_test_split(train, test_size=0.2, random_state=7759) 
```

```python
train_set
```

```python
test_set
```

```python
train_set.drop(['value'], axis = 1, inplace = True)
test_set.drop(['value'], axis = 1, inplace = True)
```

```python
X_train = train_set.drop("value_log",axis=1)
y_train = train_set["value_log"].copy()
X_val = test_set.drop("value_log",axis=1)
y_val = test_set["value_log"].copy()
X_train
```

```python
X_num = X_train.drop(["continent","position","prefer_foot"],axis=1) #스케일 하려고 수치형, 범주형 따로 빼는 작업
X_vnum = X_val.drop(["continent","position","prefer_foot"],axis=1)
X_cat = X_train[["continent","position","prefer_foot"]]
X_vcat = X_val[["continent","position","prefer_foot"]]
```

```python
from sklearn.preprocessing import StandardScaler    # 수치형 변수 -> 스텐다드 스케일러

scaler = StandardScaler()
scaler.fit(X_num)
X_scaled = scaler.transform(X_num)

vscaler = StandardScaler()
vscaler.fit(X_vnum)
X_vscaled = vscaler.transform(X_vnum)

type(X_vscaled)
```

```python
from sklearn.preprocessing import OneHotEncoder             # 범주형 변수 -> 원 핫 인코더

cat_encoder = OneHotEncoder(sparse=False)
X_cat_1hot = cat_encoder.fit_transform(X_cat)
X_cat_1hot           # 범주형 변수 -> 원 핫 인코더

cat_encoder = OneHotEncoder(sparse=False)
X_cat_v1hot = cat_encoder.fit_transform(X_vcat)
X_cat_v1hot   
```

```python
a = pd.DataFrame(X_scaled)          #  스케일 된 수치형, 변수형 합쳐서 X_train 생성 
b = pd.DataFrame(X_cat_1hot)
X_train = pd.concat([a,b],axis =1)  
X_train
```

```python
a = pd.DataFrame(X_vscaled)          #  스케일 된 수치형, 변수형 합쳐서 X_val 생성 
b = pd.DataFrame(X_cat_v1hot)
X_val = pd.concat([a,b],axis =1)  
X_val
```

```python
from sklearn.linear_model import LinearRegression

model_linear = LinearRegression()
model_linear.fit(X_train, y_train)
model_linear.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

X_predictions = model_linear.predict(X_train)

lin_mse = mean_squared_error(y_train, X_predictions, squared=True) # mean squared error
lin_rmse = mean_squared_error(y_train, X_predictions, squared=False) # root mean squared error  false는 루트값을 씌우줌
lin_mae = mean_absolute_error(y_train, X_predictions) # mean absolute error

print(lin_mse)
print(lin_rmse)
print(lin_mae)
```

```python
predicted = X_predictions
actual = y_train
act_pred_df = pd.DataFrame({'actual': actual

                            , 'predicted': predicted

                            , 'error': actual - predicted})
act_pred_df.head()
```
